```
# how to play with set add, update
def make_node(data):
    set_nodes = set()
    if len(data.split(', ')) == 1:
        set_nodes.add(data)
    elif len(data.split(', ')) > 1:
        set_nodes.update(set(data.split(', ')))
    return set_nodes
```
