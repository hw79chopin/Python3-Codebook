````python
# subplot with customizing
fig, axs = plt.subplots(1,4, figsize=(22, 8), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)
fig.tight_layout(pad=7.0)
axs = axs.ravel()

num=0
for i in df_Act['Class'].unique():
    axs[num].yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1, decimals=None, symbol='%', is_latex=False))
    axs[num].yaxis.set_tick_params(labelsize=15)
    axs[num].xaxis.set_tick_params(labelsize=18)

    df_temp = df_Act[df_Act['Class']==i]
    x_val = df_temp['OTT'].values
    y_val = df_temp['Avg.'].values
    axs[num].bar(x_val, y_val,color=['deepskyblue', 'lightblue'])

    # confidential interval
    axs[num].vlines(x='Free',
           ymin = df_temp[df_temp['OTT']=='Free']['Upper'].iloc[0],
           ymax = df_temp[df_temp['OTT']=='Free']['Lower'].iloc[0], 
           colors='black', 
           ls='-', lw=1)
    axs[num].hlines(df_temp[df_temp['OTT']=='Free']['Upper'].iloc[0],
               -0.06, 0.06, color='black', linestyle='solid', linewidth=1)
    axs[num].hlines(df_temp[df_temp['OTT']=='Free']['Lower'].iloc[0],
               -0.06, 0.06, color='black', linestyle='solid', linewidth=1)

    axs[num].vlines(x='Paid',
           ymin = df_temp[df_temp['OTT']=='Paid']['Upper'].iloc[0],
           ymax = df_temp[df_temp['OTT']=='Paid']['Lower'].iloc[0], 
           colors='black', 
           ls='-', lw=1)
    axs[num].hlines(df_temp[df_temp['OTT']=='Paid']['Upper'].iloc[0],
               0.94, 1.06, color='black', linestyle='solid', linewidth=1)
    axs[num].hlines(df_temp[df_temp['OTT']=='Paid']['Lower'].iloc[0],
               0.94, 1.06, color='black', linestyle='solid', linewidth=1)
    yticks = np.linspace(0, df_temp['Upper'].max(), num=5)
    axs[num].set_yticks(yticks)
    axs[num].set_title(i, size=25, pad=20)
    num+=1
````

````python
# tick ÎßåÎìúÎäî Ìï®Ïàò
## ÏÑ†Î∞∞Í∞Ä ÎßåÎì§Ïñ¥Ï§Ä Í∑ÄÏ∞ÆÏïÑÏÑú ÎßåÎì† Ìï®Ïàò
## ÎùºÎ≤® ÎÑ£ÏúºÎ©¥, ÎùºÎ≤® ÏàòÎßåÌÅº ÏïåÏïÑÏÑú Ï™ºÍ∞úÏ£ºÍ≥†, Ïà´Ïûê ÎÑ£ÏúºÎ©¥, Ïà´ÏûêÎåÄÎ°ú Î≤îÏúÑÎ•º Ï™ºÍ∞†Îã§Ïùå ÏÜåÏà´Ï†ê ÏïÑÎûò 2Î≤àÏß∏ ÏïÑÎûò ÍπåÏßÄ Ìï¥Îãπ Í∞íÏùÑ return
def t_generator(xy,x_lower,x_upper, labels):
    if type(labels) == int:
        label_length=labels-1
        tick_range=np.arange(x_lower,x_upper+0.00001,(x_upper-x_lower)/(label_length))
        labels=np.around(tick_range,2)
    else:
        label_length=(len(labels)-1)
        tick_range=np.arange(x_lower,x_upper+1,(x_upper-x_lower)/(label_length))
    if str(xy)=="x":
            return plt.xticks(tick_range,labels)
    if str(xy)=="y":
        return plt.yticks(tick_range,labels)
````

````python
# auto label ÎßåÎìúÎäî Ìï®Ïàò
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
````
