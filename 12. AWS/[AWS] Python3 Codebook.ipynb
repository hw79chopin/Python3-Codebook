```
# fine_one(ì¡°ê±´) ì‚¬ìš©ì‹œ í•˜ë‚˜ë§Œ ì¶œë ¥
# find(ì¡°ê±´) ì‚¬ìš©ì‹œ ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ë°ì´í„° ëª¨ë‘ë¥¼ ë³´ì—¬ì¤Œ

# ì¡°ê±´ 1: Wooju Unitedë¥¼ ì°¾ì•„ì¤˜ 
team.find_one({"team_name": "Wooju United"})

# ì¡°ê±´ 2: team_rankê°€ 2ìœ„ì¸ íŒ€
result = team.find({"team_rank":2})

for row in result:
    print(row)
    
# ì¡°ê±´ 3: team_rankê°€ 2ìœ„ ë³´ë‹¤ ë†’ì€ íŒ€ ë³´ê¸° $lt
result = team.find({"team_rank" : {"$lt" : 2}})
for row in result:
    print(row)
    
# ì¡°ê±´ 4: team_rankê°€ 2ìœ„ ì´ìƒ íŒ€ ë³´ê¸° $lte
result = team.find({"team_rank" : {"$lte" : 2}})
for row in result:
    print(row)
    
# ì¡°ê±´ 5: team_rankê°€ 2ìœ„ ë³´ë‹¤ ë‚®ì€ íŒ€ ë³´ê¸° $gt
result = team.find({"team_rank" : {"$gt" : 2}})
for row in result:
    print(row)
    
# ì¡°ê±´ 4: team_rankê°€ 2ìœ„ ë³´ë‹¤ ë‚®ì€ íŒ€ ë³´ê¸° $gte
result = team.find({"team_rank" : {"$gte" : 2}})
for row in result:
    print(row)
    
# ì¡°ê±´ 5: team_rankê°€ 2ìœ„, Mongo íŒ€ ë³´ê¸°  $and
team.find_one({"team_rank": 2, "team_name": "Mongo United"})

# ì¡°ê±´ 6: team_rankê°€ 1ìœ„ì™€ 4ìœ„ ë³´ê¸° $or
result = team.find({"$or": [{"team_rank": 1}, {"team_rank": 4}]})
for row in result:
    print(row)
    
result = team.find()

for row in result:
    print(row)
```

## ğŸ” Update Data


```
## 1. í•˜ë‚˜ì”© ìˆ˜ì • update_one
## 2. ì—¬ëŸ¬ê°œ ìˆ˜ì • update_many

# 1. í•˜ë‚˜ì”© ìˆ˜ì • update_one
# ìˆ˜ì • ë‚´ìš© : Wooju Unitedë¥¼ 1ìœ„ë¡œ ë³€ê²½ 
team.update_one({"team_name": "Wooju United"}, {"$set":{"team_rank": 1}})

# 4ìœ„ì—ì„œ 1ìœ„ë¡œ ë³€ê²½ë¨
team.find_one({"team_name": "Wooju United"})

# 2. ì—¬ëŸ¬ê°œ ìˆ˜ì • update_many
# ì¡°ê±´ ìˆœìœ„ê°€ 2ìœ„ì¸ íŒ€ë“¤ì„ 3ìœ„ë¡œ ë³€ê²½
team.update_many({"team_rank":2},{"$set": {"team_rank": 3}})

result = team.find({"team_rank":3})

for row in result:
    print(row)
    
result = team.find()

for row in result:
    print(row)
```
